```python
mean_us_r = np.mean(data_us_edge_r.values)
std_us_r = np.std(data_us_edge_r.values)
mean_v2_r = np.mean(data_v2_edge_r.values)
std_v2_r = np.std(data_v2_edge_r.values)
mean_us_r,std_us_r,mean_v2_r,std_v2_r
```

```python
# 计算每列的均值和标准差
means = data_us_edge_r.mean()
std_devs = data_us_edge_r.std()

# 创建图像
plt.figure(figsize=(10, 6))

# 绘制每个值为散点图，颜色较浅
for col in data_us_edge_r.columns:
    plt.scatter([col]*len(data_us_edge_r[col]), data_us_edge_r[col], color='lightblue', alpha=0.6, edgecolor='k')

# 绘制均值线
plt.plot(data_us_edge_r.columns, means, marker='o', linestyle='-', color='blue', label='Mean')

# 绘制误差线（均值±标准差）
plt.errorbar(data_us_edge_r.columns, means, yerr=std_devs, fmt='o', color='blue', capsize=5, label='Mean ± 1 Std Dev')

# 设置轴标签
plt.xlabel('Index (Columns)')
plt.ylabel('Values')

# 添加标题
plt.title('data_v2_edge_r   Individual Values, Mean and Standard Deviation Range per Column')

# 显示图例
plt.legend()

# 显示网格
plt.grid(True)

# 显示图像
plt.show()
```
